```python
#graphique sur evolution des prix des terres en fonction du temps 
import pandas as pd
import matplotlib.pyplot as plt

# Charger les donnes filtres
df = pd.read_csv(r"C:\Users\mejdo\OneDrive\Bureau\output.csv")

# Filtrer sur le type "Non bti"
df_non_bati = df[df["type_bien"] == "Non bti"]

# Calculer le prix moyen au m par anne (valeur foncire totale / nb de transactions)
df_prix_temps = df_non_bati.groupby("year").agg({
    "valeurfonc_sum_cod2": "sum",
    "nbtrans_cod2": "sum"
}).reset_index()

# viter les divisions par zro
df_prix_temps["prixm2_moyen"] = df_prix_temps["valeurfonc_sum_cod2"] / df_prix_temps["nbtrans_cod2"]

# Tracer le graphique
plt.figure(figsize=(10, 6))
plt.plot(df_prix_temps["year"], df_prix_temps["prixm2_moyen"], marker="o", linestyle="-", color="green")
plt.title("volution du prix moyen des terres non bties en le-de-France")
plt.xlabel("Anne")
plt.ylabel("Prix moyen au m ()")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les donnes
df = pd.read_csv(r"C:\Users\mejdo\OneDrive\Bureau\output.csv")

# --- TERRES NON BTIES ---
df_non_bati = df[df["type_bien"] == "Non bti"]
df_prix_non_bati = df_non_bati.groupby("year").agg({
    "valeurfonc_sum_cod2": "sum",
    "nbtrans_cod2": "sum"
}).reset_index()
df_prix_non_bati["prixm2_moyen"] = df_prix_non_bati["valeurfonc_sum_cod2"] / df_prix_non_bati["nbtrans_cod2"]

# --- BIENS BTIS ---
df_bati = df[df["type_bien"] == "Bti"]
df_prix_bati = df_bati.groupby("year").agg({
    "valeurfonc_sum_cod1": "sum",
    "nbtrans_cod1": "sum"
}).reset_index()
df_prix_bati["prixm2_moyen"] = df_prix_bati["valeurfonc_sum_cod1"] / df_prix_bati["nbtrans_cod1"]

# --- TRAC DES DEUX COURBES ---
plt.figure(figsize=(10, 6))

# Courbe des terres non bties
plt.plot(df_prix_non_bati["year"], df_prix_non_bati["prixm2_moyen"],
         marker="o", linestyle="-", color="green", label="Non bti")

# Courbe des biens btis
plt.plot(df_prix_bati["year"], df_prix_bati["prixm2_moyen"],
         marker="o", linestyle="-", color="blue", label="Bti")

# Personnalisation
plt.title("volution du prix moyen au m en le-de-France")
plt.xlabel("Anne")
plt.ylabel("Prix moyen au m ()")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
